### Tourist site details
#### <span style="color:blue;"> The function below will get all the Tourist site names, classifications, and Other data from the given URL 

```python
# Scrapping Tourist sites Name, classification, and number of reviews
def scrape_toursite(url_x):    
    service = Service(executable_path=CHKS) 
    options = webdriver.EdgeOptions()
    options.add_argument("--headless")
    driver = webdriver.Edge(service=service, options=options)

    # Use Selenium to open the webpage
    driver.get(url_x)

    # Number of times to scroll down
    num_scrolls = random.randint(9,10)

    # Delay between each scroll action (in seconds)
    scroll_delay = random.randint(5,6)  # Adjust as needed
    
    time.sleep(random.randint(5,6)) # implementing sleep time

    # Scrolling
    for i in range(num_scrolls):
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)
        time.sleep(scroll_delay)

    # Retrieve the entire page source
    page_source = driver.page_source

    # Parse the page source with BeautifulSoup
    soup = BeautifulSoup(page_source, 'html.parser')
    
    toursite_list = []

    tag_lst = ['div',  # Toursite name
               'div',  # Classfication
               'span'] # Number of reviews
    class_lst = ['XfVdV o AIbhI', # Toursite name
                'alPVI eNNhq PgLKC tnGGX yzLvM',# Classfication
                'biGQs _P pZUbB osNWb'] # Number of reviews

    for tag, clss in zip(tag_lst, class_lst):
        elements = soup.find_all(tag, class_=clss)
        if tag == 'div' and clss == 'alPVI eNNhq PgLKC tnGGX yzLvM':
            for ele in elements:
                toursite_list.append(ele.find_next('div', 'biGQs _P pZUbB hmDzD').text.strip())
        else:
            for element in elements:
                toursite_list.append(element.text.strip())
        
    driver.quit()
    
    return toursite_list
```

### Tourist site URL's
#### <span style="color:blue;"> The function below will get all the Tourist site URLs, these URL's will be used later for scrapping location and reviews

```python
# This will extract the URLs of tourist site
def get_urls(source_page):
    service = Service(executable_path=CHKS) 
    options = webdriver.EdgeOptions()
    options.add_argument("--headless")
    driver = webdriver.Edge(service=service, options=options)

    # Use Selenium to open the webpage
    driver.get(source_page)

    # Number of times to scroll down
    num_scrolls = random.randint(9,10)

    # Delay between each scroll action (in seconds)
    scroll_delay = random.randint(5,6)  # Adjust as needed
    
    time.sleep(random.randint(5,6))
    
    # Perform scrolling
    for i in range(num_scrolls):
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)
        time.sleep(scroll_delay)

    # Retrieve the entire page source
    page_source = driver.page_source

    # Parse the page source with BeautifulSoup
    soup = BeautifulSoup(page_source, 'html.parser')

    # Find the tourist site name element by class and attribute
    name_element = soup.find_all('a', class_='BUupS')

    # Close the browser
    driver.quit()
    
    url_listing = []
    
    # Get the href attribute
    for elements in name_element:
        href_attribute = elements.get('href')    
        if href_attribute and href_attribute.startswith('/'): # Check if the href starts with '/', which indicates it's a relative URL
            full_url = 'https://www.tripadvisor.com.ph' + href_attribute # Build the full URL by prepending 'https://www.tripadvisor.com.ph'
            url_listing.append(full_url)
        else:
            url_listing.append(href_attribute) # If the href is an absolute URL, use it directly
            
    return url_listing
    
```

```python
# Scrapping Tourist sites locations
def scrape_toursite_loc(url_loc_x):
    service = Service(executable_path=CHKS)
    options = webdriver.EdgeOptions()
    options.add_argument("--headless")
    driver = webdriver.Edge(service=service, options=options)

    # Use Selenium to open the webpage
    driver.get(url_loc_x)

    # Number of times to scroll down
    num_scrolls = 2

    # Delay between each scroll action (in seconds)
    scroll_delay = random.randint(5,6)  # Adjust as needed
    
    time.sleep(random.randint(5,6))
    
        # To bypass pop up cart
    try:
        # Check if the specific div element exists
        if driver.find_element(By.XPATH, '/html/body/div[3]/div[2]/div[2]'):
            try:
                # Find the body element
                body_element = driver.find_element(By.TAG_NAME, 'body')
                body_element.click()
            except NoSuchElementException:
                pass
    except NoSuchElementException:
        pass
    
    time.sleep(random.randint(2,3))

    # Perform scrolling
    for i in range(num_scrolls):
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)
        time.sleep(scroll_delay)

    # Retrieve the entire page source
    page_source = driver.page_source

    # looking for a specific element using xpath function in selenium 
    toursite_elem_loc = []
    try:
        if driver.find_element(By.XPATH, '/html/body/div[1]/main/div[1]/div[2]/div[2]/div[2]/div/div[1]/section[5]/div/div/div[2]/div[1]/div/div/div/div/button/span'):                                                
            soup = BeautifulSoup(page_source, 'html.parser')
            loc_element = soup.find_all('span', class_='biGQs _P XWJSj Wb') 
            # this will eliminate all unnecessary tags, we only want the tag with Philippines on it.
            for element in loc_element:
                for item in element:
                    if 'Philippines' in item:
                        toursite_elem_loc.append(element.text.strip())
    except NoSuchElementException:
        toursite_elem_loc.append('')


    # Close the browser
    driver.quit()
            
    return toursite_elem_loc
```

```python
# scrapper function for missing location using bing maps
def get_miss_loc(site):
    service = Service(executable_path=CHKS)
    options = webdriver.EdgeOptions()
    options.add_argument("--headless")
    driver = webdriver.Edge(service=service, options=options)

    # Use Selenium to open the webpage
    driver.get('https://www.bing.com/maps/?cp=10.955322%7E123.892162&lvl=8.1')

    name_element = []
    element = []
    
    # Check if the specific search bar element exists
    search_bar = driver.find_element(By.XPATH, '//*[@id="maps_sb"]')
    if site == 'Pescador Island':
        search_bar.send_keys(site + 'Boats') # need to add keyword as this place is an island
    else:
         search_bar.send_keys(site)
    time.sleep(1) # rest per action

    if driver.find_element(By.XPATH, '//*[@id="sb_as_container"]'):
        search_bar.send_keys(Keys.ARROW_DOWN) # pressing enter
        time.sleep(random.randint(3,4))
        search_bar.send_keys(Keys.RETURN)

    time.sleep(random.randint(2,3))  
    current_url = driver.current_url # retrive the new url
    time.sleep(random.randint(2,3))
    #driver.get(current_url)
    time.sleep(random.randint(2,3))
    page_source = driver.page_source  # Retrieve the entire page source
    time.sleep(random.randint(2,3))
    soup = BeautifulSoup(page_source, 'html.parser') # Parse the page source with BeautifulSoup
    name_element.append(soup.find_all('div', class_='iconDataList')) # Find the missing location
    time.sleep(random.randint(2,3))
    
    driver.quit() # Close the browser

    for elements in name_element:
        for ele in elements:
            element.append(ele.text.strip())

    return element
    
```

```python
# Scrapping Tourist sites reviews
def scrape_toursite_rev(url_x):    
    service = Service(executable_path=CHKS) 
    options = webdriver.EdgeOptions()
    options.add_argument("--headless")
    driver = webdriver.Edge(service=service, options=options)

    # Use Selenium to open the webpage
    driver.get(url_x)

    # Number of times to scroll down
    num_scrolls = random.randint(9,10)

    # Delay between each scroll action (in seconds)
    scroll_delay = random.randint(5,6)  # Adjust as needed
    
    time.sleep(random.randint(5,6))
    
        # To bypass pop up cart
    try:
        # Check if the specific div element exists
        if driver.find_element(By.XPATH, '/html/body/div[3]/div[2]/div[2]'):
            try:
                # Find the body element
                body_element = driver.find_element(By.TAG_NAME, 'body')
                body_element.click()
            except NoSuchElementException:
                pass
    except NoSuchElementException:
        pass
    
    time.sleep(random.randint(2,3))

    # Perform scrolling
    for i in range(num_scrolls):
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)
        time.sleep(scroll_delay)

    # Retrieve the entire page source
    page_source = driver.page_source

    # Parse the page source with BeautifulSoup and Find the tourist site name element by class and attribute
    if driver.find_element(By.XPATH, '//*[@id="lithium-root"]/main/div[1]/div[2]/div[2]/div[2]/div/div[1]/section[6]/div/div/div/div[2]/div/div/div/ul/li[1]/div[1]/div[1]/div[1]/div/div[2]/span/a') or driver.find_element(By.XPATH, '//*[@id="lithium-root"]/main/div[1]/div[2]/div[2]/div[2]/div/div[1]/section[6]/div/div/div/div[2]/div/div/div/ul/li[1]/div[1]/div[4]/div[1]/div/span/div'):                                                
        soup = BeautifulSoup(page_source, 'html.parser')
        name_element = soup.find_all(['span', 'span'], class_=['biGQs _P fiohW fOtGX','JguWG']) 
        touriste_elem_list = [element.text.strip() for element in name_element]

    # Close the browser
    driver.quit()
            
    return touriste_elem_list
```
