### For restaurant details
#### <span style="color:blue;"> The function below will get all the Restaurant names, Location, Specialty, Price and Other data from the given URL 

```python
# Scrapping Restaurant Name, location, price, specialty, and other data
def scrape_restau(url_x):    
    edge_service = Service(CHKS)
    edge_options = Options()
    edge_options.add_argument("--headless")
    driver = webdriver.Edge(service=edge_service, options=edge_options)

    # Use Selenium to open the webpage
    driver.get(url_x)

    # Number of times to scroll down
    num_scrolls = 2

    # Delay between each scroll action (in seconds)
    scroll_delay = random.randint(5,6)  # Adjust as needed
    
    time.sleep(random.randint(3,4))

    # Perform scrolling
    for i in range(num_scrolls):
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_UP)
        time.sleep(scroll_delay)
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)
        time.sleep(scroll_delay)

    # Retrieve the entire page source
    page_source = driver.page_source

    # Parse the page source with BeautifulSoup
    soup = BeautifulSoup(page_source, 'html.parser')


    # Find the restaurant name element by class and attribute
    tag_lst = ['h1', # restaurant name
               'a',  # location
               'div', # price range and specialty outer div
               'span', # overall_ratings
               'a'] # count_reviews
    class_lst = ['HjBfq', # restaurant name
                 'AYHFM', # location
                 'tbUiL', # price range and specialty outer class
                 'ZDEqb', # overall_ratings
                 'IcelI'] # count_reviews

    alter_tag = ['div']  # price range and specialty
    alter_clss = ['CtTod Wf b'] # price range and specialty
    rem_list = ['Website', 'Menu']
    x_list = []
    
    # Close the browser
    driver.quit()
    
    # Evaluate and extract html element
    for tag, clss in zip(tag_lst, class_lst):
        elements = soup.find_all(tag, class_=clss)
        if tag == 'div' and clss == 'tbUiL' and len(elements) == 0:
            alter_elements = soup.find_all(alter_tag, class_=alter_clss)
            x_list.insert(2, '')
            x_list.insert(3, '')
            for element in alter_elements:
                inner_elem = element.find_next('div', class_='AGRBq').text.strip()
                if element.text.strip() == 'PRICE RANGE':
                    x_list[2] = inner_elem
                elif element.text.strip() == 'CUISINES':
                    x_list[3] = inner_elem
        elif tag == 'div' and clss == 'tbUiL':
            x_list.insert(2, '')
            x_list.insert(3, '')
            for element in elements:
                inner_elem = element.find_next('div', class_='SrqKb').text.strip()
                if element.text.strip() == 'PRICE RANGE':
                    x_list[2] = inner_elem
                elif element.text.strip() == 'CUISINES':
                    x_list[3] = inner_elem
        else:
            for element in elements:
                if element.text.strip() not in set(rem_list) and '#' not in element.text.strip():
                    x_list.append(element.text.strip())
   
    return x_list
```


### Scrapping retaurant reviews
#### <span style="color:blue;"> This function will scrapped 15 reviews from the restaurant we previously get data from

```python
# Scrapping Restaurant reviews
def scrape_restau_rev(url_x):    
    service = Service(executable_path=CHKS)
    options = webdriver.EdgeOptions()
    options.add_argument("--headless")
    driver = webdriver.Edge(service=service, options=options)

    # Use Selenium to open the webpage
    driver.get(url_x)

    # Number of times to scroll down
    num_scrolls = random.randint(8,9)

    # Delay between each scroll action (in seconds)
    scroll_delay = random.randint(3,4)  # Adjust as needed
    
    time.sleep(random.randint(3,4))

    # Perform scrolling
    for i in range(num_scrolls):
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)
        time.sleep(scroll_delay)
        
    # search for the 'more' button    
    try:
        if driver.find_element(By.XPATH, '//*[@id="review_913594862"]/div/div[2]/div[2]/div/p/span[2]'): 

            try:
                btn = driver.find_element(By.CLASS_NAME, 'taLnk ulBlueLinks') 
                btn.click() # to click  'more' button
            except NoSuchElementException:
                pass

    except NoSuchElementException:
        pass


    # Retrieve the entire page source
    page_source = driver.page_source

    # Parse the page source with BeautifulSoup
    soup = BeautifulSoup(page_source, 'html.parser')

    # Find the restaurant name element by class and attribute
    name_element = soup.find_all(['div', 'p'], class_=['info_text pointer_cursor',
                                                       'partial_entry'])

    resto_elem_list = [element.text.strip() for element in name_element]

    # Close the browser
    driver.quit()
            
    return resto_elem_list
```
