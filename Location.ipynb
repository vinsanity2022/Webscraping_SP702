## Grouping
### <span style="color:blue;"> Now were gonna group them based on their City/Municipality/Province.</span> 

We will <span style="color:Red;"><strong>first group them by City/Municipality</strong></span>, and if they do not belong to any of these areas, we will <span style="color:Red;"><strong>then group them by Province</strong></span>

Priority order of grouping:
1. City / Municipality (1st)
2. Province (2nd)


####  <span style="color:blue;">Below are some functions that we will be using to edit the dataframe</span> 

```python
# This function will filter the dataframe based on location
def filter_loc(df, filter_lst):
    
    filtered_lst = df[df['Location'].str.contains('|'.join(filter_lst), case=False, na=False)]
    
    return filtered_lst
```

```python
# This function will change the column name for hotel, restaurant, and tourist site.
# it will also add the 'Category' column, where we will incorporate some form of identification based on the 'Place_name'
def format_name_col(df, rep_name):
    
    if df.columns[0] == 'Hotel_name':
        df.insert(1,'Category', 'Hotel')
    elif df.columns[0] == 'Restaurant_name':
        df.insert(1,'Category', 'Restaurant')
    else:
        df.insert(1, 'Category', 'Tourist site')
 
    df.columns = df.columns.str.replace(rep_name, 'Place_names')
    
    return df.columns    
```

```python
# This fucntion will recieved the dataframe and call filter_loc & format_name_col function
def edit_df(df, filter_lst, rep_name):
    
    filtered_lst = filter_loc(df, filter_lst)
      
    filtered_lst.columns = format_name_col(filtered_lst, rep_name)
    
    return filtered_lst
    
```

```python
# This function will recieved a list and create a merged data frame
def merged_data(df_lst):
    
    data_df = pd.DataFrame()
    
    for df in df_lst:
        if data_df.empty:
            data_df = df.copy()  # Initialize data_df with the first DataFrame
        else:
            common_columns = set(data_df.columns) & set(df.columns)

            if len(common_columns) > 0:
                # If there are common columns, stack the DataFrames vertically
                data_df = pd.concat([data_df, df], ignore_index=True, axis=0)
            else:
                # If there are no common columns, append the DataFrame horizontally
                data_df = pd.concat([data_df, df], axis=1)

    return data_df
```

### <span style="color:blue;"> Cebu City (Hotel, Restaurant, and Tourist site)

```python
df_cebcty_list = []

ceb_filter = ['Cebu City 6000 Philippines', 'Cebu City 6004 Philippines', 'Cebu City']
rep_list = ['Hotel_name', 'Restaurant_name', 'Tourist_site']
df_list = [df_hotel, df_resto, df_toursite]
i=0

# this for loop will call the function edit_df to filter dataframe listed in df_list 
# and change the name of a column in rep_list.
for df in df_list:
    df_cebcty_list.append(edit_df(df, ceb_filter, rep_list[i]))
    i+=1

# call the function merge_data
cebcty = merged_data(df_cebcty_list)
cebcty
```

```python
# Saving the dataframe to csv file
cebcty.to_csv('Location_Cebu.csv', index=False)
```

### <span style="color:blue;">Summary</span>
#### <span style="color:blue;"> These are the Categories with their respective counts</span>
#### <span style="color:blue;"> Below is a function that will count each category (hotel, restaurant, and Tourist site)</span>

```python
# will count the values based on the category
def summ_out(df, cate):
    list_cate = []
    
    category_counts = df['Category'].value_counts().reindex(cate, fill_value=0)
    list_cate.extend([category_counts.iloc[0], category_counts.iloc[1], category_counts.iloc[2]])
    
    return list_cate
```

```python
categories_to_count = ['Hotel', 'Restaurant', 'Tourist site']

# Counting all category in Category column
cat_cebcty = summ_out(cebcty, categories_to_count)
cat_lapcty = summ_out(lapcty, categories_to_count)
cat_moalmun = summ_out(moalmun, categories_to_count)
cat_mancty = summ_out(mancty, categories_to_count)
cat_prov = summ_out(prov, categories_to_count)
```
